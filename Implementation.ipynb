```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings ('ignore')
```

```python
df=pd.read_csv('Diabetic_Prediction.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.info()
```

```python
x=df.drop('diabetes', axis=1)
x
```

```python
y=df['diabetes']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
rf=RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Lr:", accuracy_score(y_test, lr.predict(x_test)))
print("Dt:", accuracy_score(y_test, dt.predict(x_test)))
print("Rf:", accuracy_score(y_test, rf.predict(x_test)))
print("Knn:", accuracy_score(y_test, knn.predict(x_test)))
```

```python
rf=RandomForestClassifier(
    n_estimators=100, random_state=42, class_weight='balanced'
)
rf.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred=rf.predict(x_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score
y_prob=rf.predict_proba(x_test)[:,1]
print("ROC AUC Score:", roc_auc_score(y_test, y_prob))
```
